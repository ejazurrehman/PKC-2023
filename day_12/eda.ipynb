### 3. Finding Missing Values

```python
df.isnull().sum ()
```

```python
# Calculate missing value percentage
missing_percentage = round((df.isnull().sum() / len(df)) * 100, 2)
# Display missing value percentages
print(missing_percentage)
# Visualize missing value percentages
missing_percentage.plot(kind='bar', figsize=(8, 6), color='blue')
plt.title('Missing Value Percentage in Titanic Dataset')
plt.xlabel('Columns')
plt.ylabel('Percentage of Missing Values')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a heatmap for missing values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=True, cmap='viridis', yticklabels=False)
plt.title('Missing Values Heatmap')
plt.show()
```

### 4. Handling Missing Values

##### 4.1 Age Column (Numeric Data)

```python
# Fill missing values in 'Age' with the median
df['age'].fillna(df['age'].median(), inplace=True)

# Verify
print(df['age'].isnull().sum())  # Should output 0
```

##### 4.2 Embarked Column (Categorical Data)

```python
# Fill missing values in 'Embarked' with the mode
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)

# Verify
print(df['embarked'].isnull().sum())  # Should output 0
```

##### 4.3 Embarked Town Column (Categorical Data)

```python
# Fill missing values in 'Embarked' with the mode
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)

# Verify
print(df['embark_town'].isnull().sum())  # Should output 0
```

##### 4.4 Deck Column (Categorical Data)

```python
# Drop the 'deck' column
df.drop('deck', axis=1, inplace=True)

# Verify
print('deck' in df.columns)  # Should output False
```

```python
# Check for missing values again
print(df.isnull().sum())
```

```python
# Plot the distribution of 'Age' before and after filling missing values
plt.figure(figsize=(12, 6))

# Before handling missing values
plt.subplot(1, 2, 1)
sns.histplot(df['age'].dropna(), kde=True, color='blue')
plt.title('Age Distribution (Before Imputation)')

# After handling missing values
plt.subplot(1, 2, 2)
sns.histplot(df['age'], kde=True, color='green')
plt.title('Age Distribution (After Imputation)')

plt.show()
```

#### 5.6 Calculate Outliers via IQR Method

```python
# To calculate outliers in the Titanic dataset using Python, we can use the Interquartile Range (IQR) method. 
# Outliers are typically defined as data points that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR, where:

# Q1 is the first quartile (25th percentile),

# Q3 is the third quartile (75th percentile),

# IQR is the interquartile range (Q3 - Q1).
# Calculate Q1, Q3, and IQR for the 'age' column
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers in the 'age' column
outliers = df[(df['age'] < lower_bound) | (df['age'] > upper_bound)]

# Print results
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound for Outliers: {lower_bound}")
print(f"Upper Bound for Outliers: {upper_bound}")
print("\nOutliers in the 'age' column:")
print(outliers[['age']])

# Visualize outliers using a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['age'], color='lightblue')
plt.title('Boxplot of Age with Outliers', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.show()
```

#### 5.8 Remove Outliers from the Data

```python
# Remove Outliers from the data

# Calculate Q1, Q3, and IQR for the 'age' column
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the 'age' column
df_cleaned = df[(df['age'] >= lower_bound) & (df['age'] <= upper_bound)]

# Print results
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound for Outliers: {lower_bound}")
print(f"Upper Bound for Outliers: {upper_bound}")
print(f"\nOriginal dataset shape: {df.shape}")
print(f"Cleaned dataset shape: {df_cleaned.shape}")

# Visualize the cleaned data
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_cleaned['age'], color='lightblue')
plt.title('Boxplot of Age (Outliers Removed)', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.show()
```

#### 5.10 Stacked Bar Plot

```python
# Stacked Bar Plot of Survival by Class

# Calculate survival counts per class
survival_counts = df.groupby(['class', 'survived']).size().unstack()
survival_counts.columns = ['Not Survived', 'Survived']  # Rename columns
# Plot
survival_counts.plot(kind='bar', stacked=True, color=['#ff7f0e', '#1f77b4'], figsize=(8, 6))
plt.title('Survival by Passenger Class', fontsize=16)
plt.xlabel('Class', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='Survival Status')
plt.xticks(rotation=0)
plt.show()
```

```python
# Stacked Bar Plot of Survival by Class and Gender

# Calculate survival counts by class and gender
survival_gender_counts = df.groupby(['class', 'sex', 'survived']).size().unstack([1, 2])
survival_gender_counts.columns = ['Female Not Survived', 'Female Survived', 'Male Not Survived', 'Male Survived']

# Reorder columns for logical stacking
survival_gender_counts = survival_gender_counts[['Male Not Survived', 'Male Survived', 'Female Not Survived', 'Female Survived']]
# Define colors for each segment
colors = ['#d62728', '#2ca02c', '#ff7f0e', '#1f77b4']

# Plot
ax = survival_gender_counts.plot(kind='bar', stacked=True, color=colors, figsize=(8, 6))
plt.title('Survival by Class and Gender', fontsize=16)
plt.xlabel('Class', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.show()
```

```python
# a scatter plot with a regression line, showing how fare and age relate for passengers who paid less than 300

# Filter the dataset where fare is less than 300
df_filtered = df[df['fare'] < 300]

# Draw the LM plot
sns.lmplot(data=df_filtered, x="age", y="fare", hue="sex", aspect=1.5)

# Show the plot
plt.title("LM Plot of Age vs Fare (Fare < 300)")
plt.show()
```

```python
# lm plot after removal of outliers

# Filter dataset for fare < 300
df_filtered = df[df["fare"] < 300]

# Function to remove outliers using IQR method
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)  # First quartile
    Q3 = df[column].quantile(0.75)  # Third quartile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers for age and fare
df_filtered = remove_outliers(df_filtered, "age")
df_filtered = remove_outliers(df_filtered, "fare")

# Plot LM plot
sns.lmplot(data=df_filtered, x="age", y="fare", hue="sex", aspect=1.5)

# Show the plot
plt.title("LM Plot of Age vs Fare (Fare < 300, Outliers Removed)")
plt.show()
```

Summary of Findings: Survival Rate by Family Size on Titanic
The analysis of survival rates based on family size in the Titanic dataset reveals key insights into how family dynamics influenced survival outcomes. The data suggests a clear relationship between family size and survival probability.

Small Families Had Higher Survival Rates:

Passengers traveling with one or two family members (family size 2-3) had the highest survival rates.
This suggests that smaller families had a better chance of staying together and accessing lifeboats.
Solo Travelers Faced Lower Survival Chances:

Individuals traveling alone (family size 1) had a significantly lower survival rate.
This may indicate difficulty in securing a spot on lifeboats without family support.
Larger Families Had Reduced Survival Rates:

Families with four or more members showed a decline in survival rates.
This could be due to logistical challenges in evacuating together or prioritization of smaller groups during rescue efforts.
Conclusion:
The findings highlight the impact of family structure on survival, with small family groups benefiting the most. The lower survival rate for solo travelers and large families indicates that social connections played a crucial role in survival decisions during the disaster.
