# 9. Converting string into number

```python
df = pd.DataFrame ({'col_a': ['1','2','3','8','9','10','11','12'], 'col_b': ['4','5','6','13','14','15','16','17']})
df
```

```python
df.dtypes
```

```python
df.astype({'col_a': 'int64' , 'col_b': 'int64' })
df.dtypes
```

```python
pd.to_numeric (df['col_a'], errors='coerce')
```

```python
df.dtypes
```

```python
pd.to_numeric (df['col_b'], errors='coerce')
```

```python
df.dtypes
```

```python
# again both columns shows as object type. The issue arises because pd.to_numeric() returns a transformed Series but does not modify
# the DataFrame in place. Our commands convert the data types for col_a and col_b temporarily but do not assign the results back to the
# DataFrame. To fix this issue, we need to explicitly assign the transformed data back to the columns in the DataFrame. 
# Here's how you can do it in the following way.

df['col_a'] = pd.to_numeric(df['col_a'], errors='coerce')
df['col_b'] = pd.to_numeric(df['col_b'], errors='coerce')
```

```python
df.dtypes
```
