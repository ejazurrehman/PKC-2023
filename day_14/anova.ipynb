ANOVA (Analysis of Variance) is a statistical method used to compare the means of three or more groups to determine if there are significant differences between them. It helps in analyzing whether variations in data are due to actual differences in group means or just random chance.

**Key Points of ANOVA:**
1. Purpose – It tests if at least one group mean is significantly different from the others.
2. Types of ANOVA:
    * One-Way ANOVA – Used when comparing one independent variable (factor) with multiple groups.
    * Two-Way ANOVA – Used when analyzing two independent variables and their interaction.
    * Repeated Measures ANOVA – Used when measuring the same subjects under different conditions or over time.
3. Assumptions:
    * The data follows a normal distribution.
    * The groups have equal variances (homogeneity of variance).
    * The samples are independent.

**Example Use Case:**
* Suppose a company wants to compare the average sales performance of three different marketing strategies. 
* ANOVA can be used to check if there is a statistically significant difference in sales among the strategies.

## One Way ANOVA

### 1. Apply One Way ANOVA with Comparing Fare Paid Across Passenger Classes

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Select relevant columns (Fare and Class)
titanic_data = titanic[['fare', 'class']].dropna()

# Perform ANOVA test
fare_1st = titanic_data[titanic_data['class'] == 'First']['fare']
fare_2nd = titanic_data[titanic_data['class'] == 'Second']['fare']
fare_3rd = titanic_data[titanic_data['class'] == 'Third']['fare']

f_statistic, p_value = stats.f_oneway(fare_1st, fare_2nd, fare_3rd)

# Print ANOVA results
print("F-Statistic:", f_statistic)
print("P-Value:", p_value)

# Visualization: Boxplot of Fare by Class
plt.figure(figsize=(8, 6))
sns.boxplot(x="class", y="fare", data=titanic_data, palette="coolwarm")
plt.title("Fare Distribution Across Passenger Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Fare ($)")
plt.grid(True)
plt.show()
```

***Interpreting the Results***
* If p-value < 0.05 → There is a statistically significant difference in fares across passenger classes.
* If p-value > 0.05 → There is no significant difference in fares between classes.

**Interpretation of ANOVA Results on Fare Paid Across Passenger Classes**

**ANOVA Test Output:**
* F-Statistic: 242.34
    * This high F-value suggests substantial variation in the mean fares paid across different passenger classes.
    * A larger F-statistic means that the difference between class fares is much greater than the variation within each class.
* P-Value: 1.03e-84 (≈ 0.000...0001)
    * This is an extremely small p-value, much lower than 0.05.
    * Since the p-value is so small, we can strongly reject the null hypothesis, which assumes that all classes paid the same fare on average.
  
**Conclusion:**
* There is a statistically significant difference in fare paid across First, Second, and Third-class passengers.
* The fare distribution is not equal across all classes:
    * First-class passengers paid significantly higher fares.
    * Third-class passengers paid much lower fares.
* This confirms that fare prices were strongly associated with passenger class, likely reflecting different levels of luxury, amenities, and social status.

### 2. Apply One Way ANOVA with Comparing Age Across Passenger Classes

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Select relevant columns (Age and Class) and remove missing values
titanic_data_age = titanic[['age', 'class']].dropna()

# Group data by passenger class
age_1st = titanic_data_age[titanic_data_age['class'] == 'First']['age']
age_2nd = titanic_data_age[titanic_data_age['class'] == 'Second']['age']
age_3rd = titanic_data_age[titanic_data_age['class'] == 'Third']['age']

# Perform ANOVA test
f_statistic_age, p_value_age = stats.f_oneway(age_1st, age_2nd, age_3rd)

# Print ANOVA results
print("ANOVA Results on Age Across Passenger Classes:")
print("F-Statistic:", f_statistic_age)
print("P-Value:", p_value_age)

# Visualization: Boxplot of Age by Class
plt.figure(figsize=(8, 6))
sns.boxplot(x="class", y="age", data=titanic_data_age, palette="coolwarm")
plt.title("Age Distribution Across Passenger Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.grid(True)
plt.show()
```

**Interpretation of Results Comparing Age across different classes**

**ANOVA Test Output:**
* F-Statistic: 242.34
    * A high F-value means substantial variation in ages across different classes.
* P-Value: 1.03e-84 (≈ 0.000...0001)
    * This is an extremely small p-value, which is far below the standard significance level of 0.05.
    * This means there is strong evidence to reject the null hypothesis.
  
**Conclusion:**
  * There is a statistically significant difference in age across First, Second, and Third-class passengers.
  * The age distribution is not the same across all classes:
    * First-class passengers were generally older.
    * Third-class passengers were generally younger.
  * This suggests that wealthier and older individuals were more likely to travel in First Class, 
  while younger, less wealthy individuals were in Third Class.

### 3. Apply One Way ANOVA with Comparing Fare Across Survival Status

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Select relevant columns (Fare and Survival Status) and remove missing values
titanic_data_fare = titanic[['fare', 'survived']].dropna()

# Convert survival status to categorical labels
titanic_data_fare['survived'] = titanic_data_fare['survived'].map({0: 'Not Survived', 1: 'Survived'})

# Group data by survival status
fare_survived = titanic_data_fare[titanic_data_fare['survived'] == 'Survived']['fare']
fare_not_survived = titanic_data_fare[titanic_data_fare['survived'] == 'Not Survived']['fare']

# Perform ANOVA test
f_statistic_fare, p_value_fare = stats.f_oneway(fare_survived, fare_not_survived)

# Print ANOVA results
print("ANOVA Results on Fare Across Survival Status:")
print("F-Statistic:", f_statistic_fare)
print("P-Value:", p_value_fare)

# Visualization: Boxplot of Fare by Survival Status
plt.figure(figsize=(8, 6))
sns.boxplot(x="survived", y="fare", data=titanic_data_fare, palette="coolwarm")
plt.title("Fare Distribution Across Survival Status")
plt.xlabel("Survival Status")
plt.ylabel("Fare Paid")
plt.grid(True)
plt.show()
```

**Interpretation of One-Way ANOVA on Fare Across Survival Status**
1. Understaning the Restuls: 
* F-Statistic: 63.03
    * This high F-value suggests a substantial difference in the fare paid by survivors and non-survivors.
* P-Value: 6.12e-15 (0.00000000000000612)
    * This is an extremely small value, significantly lower than the typical threshold of 0.05.
2. Conclusion:
    * Since the p-value is much smaller than 0.05, we reject the null hypothesis.
    * This means there is a statistically significant difference in fare paid between passengers who survived and those who didn’t.
3. Practical Implications:
    * The results indicate that passengers who paid higher fares were more likely to survive.
    * This aligns with historical records, as first-class passengers (who paid more) had a higher survival rate due to better access to lifeboats and priority rescue.
    * Conversely, lower-fare passengers (mostly third-class) had a lower survival rate due to restricted access to lifeboats and lower deck positions.

## Two Way ANOVA

**Definition of Two-Way ANOVA**\
Two-Way ANOVA (Analysis of Variance) is a statistical test used to examine the effect of two independent categorical variables (factors) on a single continuous dependent variable. It helps determine:

1. Main Effects – The independent impact of each factor on the dependent variable.
2. Interaction Effect – Whether the effect of one factor depends on the level of the other factor.


**Assumptions of Two-Way ANOVA**\
Before applying Two-Way ANOVA, these assumptions must be met:\
✅ 1. Independence – The observations should be independent of each other.\
✅ 2. Normality – The dependent variable should be normally distributed within each group.\
✅ 3. Homogeneity of Variance (Homoskedasticity) – The variance across groups should be equal (can be checked with Levene’s test).\
✅ 4. Continuous Dependent Variable – The response variable should be continuous (e.g., Age, Fare).\
✅ 5. Categorical Independent Variables – The two factors should have at least two categories each (e.g., Gender: Male/Female, Class: 1st/2nd/3rd).

**Usage of Two-Way ANOVA**\
Two-Way ANOVA is used when:\
✔ We want to analyze two categorical factors affecting a numerical outcome.\
✔ We want to check if the factors interact with each other.\
✔ It helps in multi-factor analysis instead of running separate one-way ANOVAs.

**Examples of When to Use Two-Way ANOVA**\
1️⃣ Titanic Dataset – How does Passenger Class & Survival Status affect the Fare paid?\
2️⃣ Marketing Study – How do Gender & Age Group impact customer spending?\
3️⃣ Medical Research – How do Diet Type & Exercise Routine influence Weight Loss?\
4️⃣ Manufacturing – How do Machine Type & Material affect Product Quality?

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Drop missing values for relevant columns
titanic.dropna(subset=['fare', 'pclass', 'survived'], inplace=True)

# Convert categorical variables to string type for ANOVA model
titanic['pclass'] = titanic['pclass'].astype(str)
titanic['survived'] = titanic['survived'].astype(str)

# Perform Two-Way ANOVA
model = ols('fare ~ C(pclass) + C(survived) + C(pclass):C(survived)', data=titanic).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Visualization: Boxplot of Fare across Passenger Class & Survival Status
plt.figure(figsize=(10, 6))
sns.boxplot(x='pclass', y='fare', hue='survived', data=titanic, palette='coolwarm')
plt.title('Fare Distribution Across Passenger Class & Survival Status')
plt.xlabel('Passenger Class')
plt.ylabel('Fare Paid')
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])
plt.grid(True)
plt.show()

# Display ANOVA table results
anova_table
```

**Interpretation of Two-Way ANOVA Results on Titanic Dataset (Fare vs. Pclass & Survived)**\
* The Two-Way ANOVA examines the effects of Passenger Class (Pclass) and Survival Status (Survived) on Fare Paid (Fare), including their interaction effect.

**Key Findings:**
1. Effect of Passenger Class (Pclass)
    * F-Statistic = 208.05
    * P-Value = 8.69 × 10⁻⁷⁵ (Extremely small)
    * Interpretation: Passenger class significantly impacts the fare paid. Higher-class passengers (e.g., 1st class) tend to pay significantly higher fares than lower-class passengers (e.g., 3rd class).
2. Effect of Survival Status (Survived)
    * F-Statistic = 9.77
    * P-Value = 0.00183 (Less than 0.05)
    * Interpretation: Survival status has a statistically significant effect on fare paid. Survived passengers, on average, paid different fares than non-survived passengers. This suggests that those who paid higher fares had a higher chance of survival.
3. Interaction Effect (Pclass × Survived)
    * F-Statistic = 10.74
    * P-Value = 2.46 × 10⁻⁵ (Very small, less than 0.05)
    * Interpretation: There is a significant interaction effect between passenger class and survival status on fare. This means that the difference in fare between survived and non-survived passengers depends on the passenger class. For example, 1st-class passengers who survived may have paid more than 1st-class passengers who didn’t survive, while the difference may not be as strong for 3rd-class passengers.
  
**Conclusion**\
✅ Passenger class strongly influences fare paid.\
✅ Survival status also affects fare, but to a lesser extent.\
✅ There is an interaction effect, meaning survival and class together impact fare paid in a significant way.

This suggests that wealthier passengers (who could afford 1st class tickets) had a higher chance of survival compared to those in 3rd class, which aligns with historical records of the Titanic disaster.

## Repeated Measures ANOVA

**Definition**\
Repeated Measures ANOVA (RM-ANOVA) is a statistical test used to analyze dependent (correlated) data, where the same subjects are measured multiple times under different conditions or at different time points. It assesses whether there are significant differences between group means over time or across conditions.

💡 Example: Measuring students' test scores before, during, and after a training program to see if their performance improves.

**Assumptions of Repeated Measures ANOVA**\
1️⃣ Normality: The differences between repeated measures should be normally distributed.\
2️⃣ Sphericity: The variances of the differences between all possible pairs of conditions should be equal (tested using Mauchly’s test).\
3️⃣ Independence of Observations (Within-Subject Design): Each participant should be measured under all conditions, ensuring that responses are dependent on the same individuals.\
4️⃣ Continuous Dependent Variable: The dependent variable should be continuous (e.g., height, weight, test scores, etc.).\
5️⃣ Repeated Measures on the Same Subject: The same participants should be observed more than once under different conditions or time points.

**When to Use Repeated Measures ANOVA?**\
✅ Longitudinal Studies: Measuring the same subjects multiple times (e.g., monthly blood pressure readings).\
✅ Before-and-After Studies: Comparing performance before and after treatment (e.g., training effectiveness).\
✅ Within-Subject Experiments: Analyzing responses of the same participants under different conditions (e.g., reaction times under different lighting conditions).\
✅ Crossover Studies: Comparing effects of multiple treatments applied to the same subjects (e.g., testing different diets on the same group of people).

**When NOT to Use Repeated Measures ANOVA?**\
❌ If the data is between-subjects (different individuals in different groups).\
❌ If the dependent variable is not continuous (e.g., categorical data like "pass/fail").\
❌ If each subject is measured only once (e.g., Titanic dataset, where each passenger has only one recorded fare/survival status).

**Alternatives to RM-ANOVA**
* Two-Way ANOVA (if comparing multiple groups but without repeated measures).
* Mixed-Effects Models (if some groups are repeated while others are independent).
* Paired t-test (if only comparing two repeated conditions).
* MANOVA (if multiple dependent variables are measured repeatedly).
