```python
# importing liabraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# simple linear regression
from sklearn.linear_model import LinearRegression
```

```python
# importing data
df = sns.load_dataset('tips')
df.head()
```

```python
# select the features from the df
X = df[['total_bill']]
# select the target from the df
y = df['tip']
```

```python
# create a regression line plot using seaborn
sns.lmplot(x='total_bill', y='tip', data=df)
```

```python
# load the linear regression model
model = LinearRegression()
# fit the model to the data
model.fit(X, y)
```

```python
# pridect teh model for unkonwn total bill value
# total bill value = 50
model.predict([[50]])
```

```python
# write the whole script in one go
# importing liabraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
# loading the dataset
df = sns.load_dataset('tips')
# creating a linear regression model
model = LinearRegression()
# defining the features and target
X = df[['total_bill']]
y = df['tip']
# fitting the model
model.fit(X, y)
# predict the tip for a total bill of 20
model.predict([[20]])[0]  # output: 3.0
```
