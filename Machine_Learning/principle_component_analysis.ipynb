- PCA in machine learning stands for Principal Component Analysis. It’s a dimensionality reduction technique used to simplify large datasets while retaining as much variability (information) as possible.
- Imagine we have data with many features (columns) — for example, 50 variables in a dataset.
- PCA transforms these original features into a new set of features, called principal components, which are:
- Uncorrelated
- Ranked by the amount of variance (information) they capture from the original data

#### Why use PCA?
1. Reduce dimensionality: Helps in reducing computational cost and avoiding overfitting.
2. Improve visualization: Allows you to project data into 2D or 3D space for better visual understanding.
3. Remove multicollinearity: Transforms correlated variables into a set of uncorrelated ones.



#### Example use case:
- Let’s say we are analyzing customer behavior with 20 input features. PCA can reduce these to just 2 or 3 principal components while still keeping most of the important patterns. These components can then be used for clustering, classification, or visualization.

#### Key concepts:
- Covariance matrix: PCA calculates this to understand how features vary with each other.
- Eigenvectors & Eigenvalues: Used to identify principal components and how much variance each one captures.

#### Important note:
- PCA is unsupervised — it doesn't consider target labels. It only cares about the structure of the input features.

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 components for visualization
X_pca = pca.fit_transform(X)

# Convert to DataFrame
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
df_pca['target'] = y

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='target', palette='deep')
plt.title('PCA of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(labels=target_names)
plt.grid(True)
plt.show()
```
